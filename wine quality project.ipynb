```python
 ############# data processing ##########
```

```python
X = wine_dataset.drop('quality',axis=1)
```

```python
print(X)
```

```python
#### binarization labelling #####
```

```python
Y = wine_dataset['quality'].apply(lambda y_value : 1 if y_value >=7 else 0)
```

```python
print(Y)
```

```python
############# TRAINING AND TESTING #########
```

```python
X_train , X_test , Y_train , Y_test = train_test_split(X,Y,test_size=0.2 ,random_state=3)
```

```python
print(Y.shape,Y_train.shape,Y_test.shape)
```

```python
                            ###########  training #########
#################### RANDOM   FOREST CLASSIFIER  #################
```

```python
model=RandomForestClassifier()
```

```python
model.fit(X_train,Y_train)
```

```python
################# calcualting accuracy #############
```

```python
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print('ACCURACY : ',test_data_accuracy)       ######## performing great!!!!
```

```python
##### building a predictive system #########
```

```python
input_data=(7.3,0.65,0,1.2,0.065,15,21,0.9946,3.39,0.47,10)
# change inputs to numpy array
input_data_as_numpy_array=np.asarray(input_data)
# reshaping the data as we are predicting only for one instance
input_data_reshaped=input_data_as_numpy_array.reshape(1,-1)
prediction = model.predict(input_data_reshaped)
print(prediction)
if(prediction[0]==1):
    print('GOOD QUALITY OF WINE');
else :
    print('BAD QUALITY OF WINE');
```
